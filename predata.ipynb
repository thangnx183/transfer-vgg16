```python
for i in range(len(ob)):
    
    images = [f for f in listdir(path+ob[i]) if isfile(join(path+ob[i], f))]
    
    for j in range(150):
        img = cv2.imread(path+ob[i]+'/'+images[j])
        cv2.imwrite(train_path+ob[i]+'/'+ob[i]+str(j)+'.jpg', img)
        
    for j in range(150,200, 1):
        img = cv2.imread(path+ob[i]+'/'+images[j])
        cv2.imwrite(valid_path+ob[i]+'/'+ob[i]+str(j)+'.jpg', img)
    
    for j in range(200,250,1):
        img = cv2.imread(path+ob[i]+'/'+images[j])
        cv2.imwrite(test_path+ob[i]+'/'+ob[i]+str(j)+'.jpg', img)
```
